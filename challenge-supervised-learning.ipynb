## Phase 1: Logistic Regression vs. Random Forest Hypothesis (unscaled data)

The hypothesis is that random forest will be a better model than linear regression.
The random forest is less influenced by outlieirs, and there is no underlying assumption of linearity in the data.


```python
# Import required modules

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
# Read the 2019 Loans CSV file into a Pandas DataFrame

py_loans_df = pd.read_csv('Resources/2019loans.csv')
py_loans_df
```

```python
# Drop the target column and create the 2019 fratures DataFrame
X_py = py_loans_df.drop('target', axis=1)
X_py.head()
```

```python
# Read the 2020 Q1 loans CSV into a DataFrame
q1_loans_df = pd.read_csv('Resources/2020Q1loans.csv')
q1_loans_df
```

```python
# Drop the target column and create a features DataFrame
X_q1 = q1_loans_df.drop('target', axis=1) 
X_q1.head()
```

```python
# Use One-hot encoding for the entire dataframe for 2019 loan data
X_py_dummies = pd.get_dummies(X_py)
print (X_py_dummies.columns)
X_py_dummies.head()
```

```python
# Use One-hot encoding for the entire dataframe for 2020 Q1 loan data
X_q1_dummies = pd.get_dummies(X_q1)
print (X_q1_dummies.columns)
X_q1_dummies.head()
```

```python
# In 2020 Q1 data, there is a column missing.  debt_settlement_flag_Y.  
X_py_dummies['debt_settlement_flag_Y'].value_counts()
```

```python
# This column needs to be added to make sure the model runs correctly
# Add the column and set the values to 0 which reflect a negative response in the field.
X_q1_dummies['debt_settlement_flag_Y']=0
X_q1_dummies.columns
```

```python
# create the target DataFrames for both 2019 and 2020 Q1 data
y_py = py_loans_df['target']
y_q1 = q1_loans_df['target']
```

```python
# Split the 2019 data into train and test
X_py_train, X_py_test, y_py_train, y_py_test = train_test_split(X_py_dummies, y_py, random_state=1)
```

```python
# Create a lotistic regression model
classifier = LogisticRegression()
classifier
```

```python
# Fit (train) model by using the 2019 training data
classifier.fit(X_py_train, y_py_train)
```

```python
# print the 2019 training score
print(f"Training Data Score for 2019 train using logistic regression: {classifier.score(X_py_train, y_py_train)}")
```

```python
# Split the 2020 Q1 data into train and test
X_q1_train, X_q1_test, y_q1_train, y_q1_test = train_test_split(X_q1_dummies, y_q1, random_state=1)
```

```python
# Print the 2020 Q1 test score
print(f"Testing Data Score for 2020 Q1 test using logistic regression: {classifier.score(X_q1_test, y_q1_test)}")
```

```python
# Create a random forest model, fitting and training using the 2019 training data
clf = RandomForestClassifier(random_state=1, n_estimators=800).fit(X_py_train, y_py_train)
```

```python
# Print the 2019 train score
print(f"The Random Forest training score for 2019 data: {clf.score(X_py_train, y_py_train)}")
```

```python
# Print the 2020 Q1 test score
print(f"The Random Forest test score for 2020 Q1 data: {clf.score(X_q1_test, y_q1_test)}")
```

### Conclusion Phase 1

Logistic Regression
* Train Score = 0.65
* Test Score = 0.53

Random Forest
* Train Score = 1.0
* Test Score = 0.67

The hypothesis was proven to be correct in this example.  The random forest classifier provided a better outcome on the 2020 Q1 test data than the logistic regression.  

## Phase 2: Logistic Regression vs. Random Forest Hypothesis (scaled data)

Hypothesis - the random forest classifier will still be the best metho even with the scaled data due to the same factors.  
The scaling will do little to improve the scores.  

```python
# Scale the data
scaler = StandardScaler().fit(X_py_train)
X_py_train_scaled = scaler.transform(X_py_train)
X_q1_test_scaled = scaler.transform(X_q1_test)
```

```python
# Print the 2019 train score with logistic regression using scaled data
print(f"2019 train score logistic regression with scaled data: {classifier.score(X_py_train_scaled, y_py_train)}")
```

```python
# Print the 2020 Q1 test score with logistic regression using scaled data
print(f"2020 Q1 test score logistic regression with scaled data: {classifier.score(X_q1_test_scaled, y_q1_test)}")
```

```python
# Create a random forest model, fitting and training using the 2019 scaled training data
clf_scaled = RandomForestClassifier(random_state=1, n_estimators=800).fit(X_py_train_scaled, y_py_train)
```

```python
# Print the 2019 training score with random forest using scaled data
print(f"2019 training score random forest with scaled data: {clf_scaled.score(X_py_train_scaled, y_py_train)}")
```

```python
# Print the 2020 Q1 test score with random forest using scaled data
print(f"2020 Q1 test score random forest with scaled data: {clf_scaled.score(X_q1_test_scaled, y_q1_test)}")
```

### Conclusion Phase 2

Logistic Regression - Scaled
* Train Score = 0.63
* Test Score = 0.51

Random Forest
* Train Score = 1.0
* Test Score = 0.67

The hypothesis was proven to be correct in this example.  The random forest classifier had a better score than the logistic regression.

## Final Conclusions

In both the unscaled and scaled data, the random forest classifier performed slightly better than the logistic regression model.  Further analysis could determine what features are significant to the target, and if those features do or do not have a linear relationship to the target.  These results drive an opinion that there is not a strong linear relationship between the key features and the target.  

Going from unscaled data to scaled data did not have a significant impact on the score of the logistic regression model (only 0.02 point improvement with scaled data).  The scaling did not change the fundamental nature of the data, therefore did not make the data more linear for the modeling.  

The random forest model did not have any change with scaled data versus unscaled.  The decisions trees from the data did not alter enough to drive a change in the scoring with the data.  That would make sense - as the underlying choices in the decision tree modeling did not change with the scaled data.
